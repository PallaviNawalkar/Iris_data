```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
```

```python
data= datasets.load_iris()
data
```

```python
data.feature_names
```

```python
data.target_names
```

```python
data.target
```

```python
df = pd.DataFrame({'Sepal_Length':data.data[:,0],
                   'Sepal_Width':data.data[:,1],
                   'Petal_Length':data.data[:,2],
                   'Petal_Width':data.data[:,3],
                   'species':data.target})
```

```python
df
```

```python
df_0 = df[df.species==0]
df_0
```

```python
df_1 = df[df.species==1]
df_1
```

```python
df_2 = df[df.species==2]
df_2
```

```python
plt.scatter(df_0['Sepal_Length'],df_0['Sepal_Width'],color='r',marker="*")
plt.scatter(df_1['Sepal_Length'],df_1['Sepal_Width'],color='g',marker="+")
plt.scatter(df_2['Sepal_Length'],df_2['Sepal_Width'],color='b',marker='^')
```

```python
plt.scatter(df_0['Petal_Length'],df_0['Petal_Width'],color='r',marker="*")
plt.scatter(df_1['Petal_Length'],df_1['Petal_Width'],color='g',marker="+")
plt.scatter(df_2['Petal_Length'],df_2['Petal_Width'],color='b',marker='^')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['species'], axis=1)
```

```python
x
```

```python
y=df.species
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100
```

```python
y_predict= model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,plot_confusion_matrix
```

```python
accuracy_score(y_test,y_predict)*100
```

```python
result = pd.DataFrame({'Actual':y_test,'Predicted':y_predict})
result
```

```python
print(classification_report(y_test,y_predict))
```

```python
performance = confusion_matrix(y_test,y_predict)
performance
```

```python
plot_confusion_matrix(model,x_test,y_test)
plt.show()
```
